### 9. Вывод значений столбца "скорость ветра" с 5 по 20

### 10. Вывод значений столбца "temp" выше среднего


```python
mean_temp = df['temp'].mean()
high_temp_values = df[df['temp'] > mean_temp]['temp'].values
high_temp_values
```

### 11. Создание нового столбца "good_temp"


```python
df['good_temp'] = df['temp'] > mean_temp
df.head()
```

### 12. Вывод строк с пустыми значениями


```python
rows_with_null_values = df[df.isnull().any(axis=1)]
rows_with_null_values
```

### 13. Вывод первых 5 строк, удовлетворяющих условию


```python
filtered_df = df[(df['temp'] > 20) & (df['windspeed'] > 20)].head(5)
filtered_df
```

### 14. Вывод значений полей в третьей строке


```python
values_third_row = filtered_df.iloc[2].values
values_third_row
```

### 15. Сортировка DataFrame по полю "count" в обратном порядке


```python
sorted_df = df.sort_values(by='count', ascending=False)
sorted_df
```

### 16. Создание нового DataFrame на основе описания поля "season"


```python
season_mapping = {1: 'spring', 2: 'summer', 3: 'fall', 4: 'winter'}
season_df = pd.DataFrame({'season': df['season'].map(season_mapping)})
season_df
```

### 17. Соединение исходного DataFrame с новым DataFrame


```python
merged_df = pd.concat([df, season_df], axis=1)
merged_df.head()
```

### 18. Группировка по временам года и количеству


```python

grouped_df = df.groupby(['season', 'count']).size().reset_index(name='frequency')
grouped_df
```

### 19. Создание pivot_table


```python
pivot_table = df.pivot_table(index='weather', columns='humidity', values='count', aggfunc='sum')
pivot_table
```

### 20. Процентное соотношение count в зависимости от времен года


```python
percentage_by_season = df.groupby('season')['count'].sum() / df['count'].sum() * 100
percentage_by_season
```
